```python
# change dictionary to DataFrame (key to column name and value to data)
# mapping key t0 column name and value to column data 
dict = {'col1' : [1, 11], 'col2' : [2, 22], 'col3' : [3, 33]}
df_dict = pd.DataFrame(dict)
print('딕셔너리로 만든 DataFrame:\n', df_dict)
```

```python
# change DataFrame(df_dict here) to ndarray type -> using 'values'
array3 = df_dict.values
print('df_dict.values 타입:', type(array3), 'df_dict.values shape:', array3.shape)
print(array3)
print(array3.tolist()) # change DataFrame to list type (values -> tolist())
print(type(array3.tolist()))
print(df_dict.to_dict('list')) # change DataFrame to dictionary type (apply to_dict())
print(type(df_dict.to_dict('list')))
```

### 정렬, Aggregation 함수, GroupBy 적용
> sort_values()

    - by : 특정 칼럼 지정 -> 이 칼럼에 대해 정렬
    - ascending=False, True 중 선택 -> 디폴트는 true(오름차순)
    - inplace=False(default) -> 원본 DF 그대로 유지하고 정렬된 DF 결과로 return, True는 원본 DF 정렬된 상태로 바뀜

```python
titanic_sorted = titanic_df.sort_values(by=['Name'])
titanic_sorted.head(5)
```

```python
# sorting multi-column -> write list type by value
titanic_sorted_name_and_pclass = titanic_df.sort_values(by=['Pclass', 'Name'], ascending=False)
titanic_sorted_name_and_pclass.head(3)
```

```python
titanic_sorted_name_and_pclass = titanic_df.sort_values(by=['Name', 'Pclass'], ascending=False)
titanic_sorted_name_and_pclass.head(3)
```

```python
titanic_df.count() # count() returns result that doesn't consider NULL
```

```python
titanic_df[['Age', 'Fare']].mean()
```

```python
titanic_groupby=titanic_df.groupby(by='Pclass')
print(type(titanic_groupby))
titanic_groupby.count()
```

```python
titanic_groupby=titanic_df.groupby('Pclass')[['PassengerId', 'Survived']].count()
titanic_groupby
```

```python
titanic_df.groupby('Pclass')['Age'].agg([max, min])
```

```python
agg_format = {'Age':'max', 'SibSp':'sum', 'Fare':'mean'}
titanic_df.groupby('Pclass').agg(agg_format)
```

### 결손 데이터 처리하기 

```python
# check if it is missing data or not -> isna()
# replace NaN to other value -> fillna()
titanic_df.isna()
```

```python
# 결손 데이터 개수 .isna().sum()
titanic_df.isna().sum()
```

```python
# cf) 반환 값 받아야 변경된 데이터 세트 값 구할 수 있고, 혹은 inpalce=True해야 실제 데이터 세트 값이 변경됨
titanic_df['Cabin']=titanic_df['Cabin'].fillna('C000') 
titanic_df.head(3)
```

```python
titanic_df['Age'] = titanic_df['Age'].fillna(titanic_df['Age'].mean())
titanic_df['Embarked'] = titanic_df['Embarked'].fillna('S')
titanic_df.head(5)
```

```python
titanic_df.isna().sum()
```
